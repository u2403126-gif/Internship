```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
data=fetch_california_housing(as_frame=True).frame
```

```python
data.head()
```

```python
X = data.drop('MedHouseVal', axis=1)
y = data['MedHouseVal']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=38)
```

```python
model = RandomForestRegressor(n_estimators=200, random_state=38)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R-squared Score:", r2_score(y_test, y_pred))
```

```python
sample_input_data = np.array([
    [8, 20, 5, 1, 1000, 3, 34, -118]
])

predicted_value = model.predict(sample_input_data)[0]

print(f"Predicted Median House Value: ${predicted_value * 100000:.2f}")
```

```python
from xgboost import XGBRegressor
```

```python
xgb_model = XGBRegressor(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    random_state=42
)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb_model.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred_xgb)
mae = mean_absolute_error(y_test, y_pred_xgb)
mse = mean_squared_error(y_test, y_pred_xgb)

print("R2 Score:", r2)
print("MAE:", mae)
print("MSE:", mse)
```

```python
sample_input_data = np.array([[8, 20, 5, 1, 1000, 3, 34, -118]])

predicted_value = xgb_model.predict(sample_input_data)[0]
print(f"Predicted Median House Value: ${predicted_value * 100000:.2f}")
```
