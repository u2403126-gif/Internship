```
import random

words = ["apple", "tiger", "chair", "house", "plant"]
word = random.choice(words)

guessed = []
for i in range(len(word)):
    guessed.append("_")

# updated structure (matches your screenshot style)
stages = [
"""
   |
   |
=========
""",
"""
   |
   |
   O
=========
""",
"""
   |
   |
   O
   |
=========
""",
"""
   |
   |
   O
  /|
=========
""",
"""
   |
   |
   O
  /|\\
=========
""",
"""
   |
   |
   O
  /|\\
  /
=========
""",
"""
   |
   |
   O
  /|\\
  / \\
=========
"""
]

wrong = 0
used = []

print("Hangman Game")

while wrong < 6:

    print(stages[wrong])

    print("Word:", end=" ")
    for ch in guessed:
        print(ch, end=" ")
    print()

    letter = input("Enter a letter: ")

    if letter in used:
        print("Already entered")
        continue

    used.append(letter)

    found = 0
    for i in range(len(word)):
        if word[i] == letter:
            guessed[i] = letter
            found = 1

    if found == 0:
        wrong = wrong + 1
        print("Wrong guess")

    if "_" not in guessed:
        print("\nYou win! Word is:", word)
        break

if "_" in guessed:
    print(stages[wrong])
    print("You lose! Word was:", word)
```
