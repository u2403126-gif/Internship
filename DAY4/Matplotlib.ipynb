```
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,14,12,18,20]
plt.plot(x, y, marker='o')
plt.title("Line Chart")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.grid(True)
plt.show()
```

```
import matplotlib.pyplot as plt
students = ["A","B","C","D","E"]
marks = [85,70,90,65,95]
plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```
import matplotlib.pyplot as plt
students = ["A","B","C","D","E"]
marks = [85,70,90,65,95]
plt.bar(students, marks, color=["red","blue","green","orange","purple"])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```
import matplotlib.pyplot as plt
activities = ["Study", "Exercise", "Sleep", "Hobbies"]
hours = [4, 1, 7, 2]
plt.pie(hours, labels=activities, autopct='%1.1f%%')
plt.title("Daily Activities")
plt.show()
```

```
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,30,40,50]
plt.scatter(x, y)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot")
plt.show()
```

```
import matplotlib.pyplot as plt
marks = [40, 50, 60, 70, 80, 90, 60, 70]
plt.hist(marks)
plt.title("Student Marks")
plt.show()
```
